```python
import pandas as pd

df_sports = pd.read_csv("sports_articles.csv")
```

```python
df_sports = df_sports.drop(columns=["Unnamed: 0"])
df_sports
```

```python
df_sports = df_sports[df_sports["abstract"] != "No abstract available"]
df_sports
```

```python
df_sports["general_category"].value_counts()
```

```python
num_categories = 10
articles_per_category = 50
total_needed = num_categories * articles_per_category

top_categories = df_sports["general_category"].value_counts().nlargest(num_categories).index

balanced_dfs = []
for category in top_categories:
    category_df = df_sports[df_sports["general_category"] == category]
    sampled_df = category_df.sample(n=min(articles_per_category, len(category_df)), random_state=42)
    balanced_dfs.append(sampled_df)

balanced_df = pd.concat(balanced_dfs).reset_index(drop=True)
balanced_df.to_csv("balanced_small_articles.csv", index=False)
```

```python
import pandas as pd

df_small = pd.read_csv("balanced_small_articles.csv")

df_small
```

```python
df_small["general_category"].value_counts()
```

```python
import pandas as pd

df_small = pd.read_csv("balanced_small_articles.csv")
df_small["text"] = df_small["title"] + " " + df_small["abstract"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english", max_features=1000)
tfidf_matrix = vectorizer.fit_transform(df_small["text"]).toarray()
```

```python
df_small["tfidf_vector"] = [vec.tolist() for vec in tfidf_matrix]
df_small.head()
```

```python
df_small = df_small.drop(columns=["Unnamed: 0"])
df_small.to_csv("small_sports_articles.csv")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
tfidf_pca_matrix = pca.fit_transform(tfidf_matrix)
```

```python
df_small["tfidf_vector"] = [vec.tolist() for vec in tfidf_pca_matrix]
df_small.to_csv("balanced_small_articles.csv")
```

## S-BERT

```python
import pandas as pd

df_small = pd.read_csv("balanced_small_articles.csv")
df_small["text"] = df_small["title"] + " " + df_small["abstract"]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-mpnet-base-v2")

embeddings = model.encode(df_small["text"], show_progress_bar=True)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
sbert_reduced = pca.fit_transform(embeddings)
```

```python
df_small["sbert_vector"] = [vec.tolist() for vec in sbert_reduced]
df_small = df_small.drop(columns=["Unnamed: 0"])
df_small.to_csv("balanced_small_articles.csv", index=False)
```

## OpenAI

```python
import pandas as pd

df_small = pd.read_csv("balanced_small_articles.csv")
df_small["text"] = df_small["title"] + " " + df_small["abstract"]
```

```python
import json

data = df_small[["news_id", "text"]].copy()
with open("batch_job.jsonl", "a", encoding="utf-8") as file:
    for _, row in data.iterrows():
        payload = {
            "custom_id": f"{row["news_id"]}",
            "method": "POST",
            "url": "/v1/embeddings",
            "body": {
                "input": f"{row["text"]}",
                "model": "text-embedding-3-large",
                "encoding_format": "float",
                "dimensions": 1024,
            },
        }
        file.write(json.dumps(payload, ensure_ascii=False) + "\n")
```

```python
from openai import OpenAI
from dotenv import load_dotenv

import os

load_dotenv()

client = OpenAI(api_key=os.getenv("API_KEY"))

file_name = "batch_job.jsonl"

batch_file = client.files.create(file=open(file_name, "rb"), purpose="batch")

print(batch_file)
```

```python
batch_job = client.batches.create(input_file_id=batch_file.id, endpoint="/v1/embeddings", completion_window="24h")
```

```python
batch = client.batches.retrieve("batch_67bc7f33f27481908c16670fd5481070")
print(batch.status)
```

```python
file_response = client.files.content(batch.output_file_id)

file_response.write_to_file("batch_job_results.jsonl")
```

```python
import json

embeddings_file = "batch_job_results.jsonl"
embeddings_to_id = dict()
with open(embeddings_file, "r") as file:
    for line in file:
        obj = json.loads(line)
        id = obj["custom_id"]
        emb = obj["response"]["body"]["data"][0]["embedding"]
        embeddings_to_id[id] = emb
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

embedding_df = pd.DataFrame(
    list(embeddings_to_id.items()), columns=["news_id", "embeddings"]
)

embeddings_matrix = np.vstack(embedding_df["embeddings"].values)

pca = PCA(n_components=10)
reduced_embeddings = pca.fit_transform(embeddings_matrix)

embedding_df["openai_vector"] = list(reduced_embeddings)
```

```python
df = pd.read_csv("balanced_small_articles.csv")

merged_df = pd.merge(df, embedding_df[["news_id", "openai_vector"]], on="news_id")
merged_df
```
